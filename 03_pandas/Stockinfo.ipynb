```python
from iexfinance.stocks import get_historical_data
from datetime import datetime
import pandas as pd
start = datetime(2018, 12, 1)
end = datetime(2019, 1, 24)
dataset = get_historical_data("MSFT", start, end, output_format='pandas')
dataset['Returns']=dataset['close'].pct_change()
dataset['StockName']='Microsoft'
```

```python
dataset.head()
```

```python
dfgoogle = get_historical_data("GOOG", start, end, output_format='pandas')
dfgoogle['Returns']=dfgoogle['close'].pct_change()
dfgoogle['StockName']='Google'
dfgoogle.head()
```

```python
dfam = get_historical_data("AMZN", start, end, output_format='pandas')
dfam['Returns']=dfam['close'].pct_change()
dfam['StockName']='Amazon'
dfam.head()
```

```python
dfapple = get_historical_data("AAPL", start, end, output_format='pandas')
dfapple['Returns']=dfapple['close'].pct_change()
dfapple['StockName']='Apple'
dfapple.tail()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
df=dataset
sns.set()
n_data = len(dataset.close)
# Number of bins is the square root of number of data points: n_bins
n_bins = int(np.sqrt(n_data))
plt.hist(dataset.close, bins=n_bins)
plt.xlabel('Close Prices')
plt.ylabel('Count')
plt.show()
```

```python
stocks=dataset.append(dfgoogle)
print(stocks.head())
print(stocks.tail())
```

```python
stocks=stocks.append(dfam)
stocks=stocks.append(dfapple)     ## így most már négy részvényt tartalmaz a dataframe.
print(stocks.head())
print(stocks.tail())
```

```python
stocks.loc['2019-01-23']
```

```python
import matplotlib.pyplot
Apple=stocks.loc[stocks.StockName=='Apple']
M=stocks.loc[stocks.StockName=='Microsoft']
Am=stocks.loc[stocks.StockName=='Amazon']
GG=stocks.loc[stocks.StockName=='Google']
matplotlib.pyplot.plot(Apple.close, color='black', label='Apple')
matplotlib.pyplot.plot(M.close, color='red', label='Microsoft')
matplotlib.pyplot.plot(Am.close, color='gold', label='Amazon')
matplotlib.pyplot.plot(GG.close, color='brown', label='Google')
matplotlib.pyplot.xlabel('Date')
matplotlib.pyplot.ylabel('Close')
matplotlib.pyplot.xticks(rotation=45)
matplotlib.pyplot.legend(loc='upper center')
matplotlib.pyplot.style.use('fivethirtyeight')
# Display the plot
matplotlib.pyplot.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.swarmplot(x="StockName", y="close", data=stocks) 
matplotlib.pyplot.xlabel("StockName")
matplotlib.pyplot.ylabel("close")
matplotlib.pyplot.show()
```

```python
import seaborn as sns
sns.boxplot(x="StockName", y="close", data=stocks)
matplotlib.pyplot.ylabel("Close price")
matplotlib.pyplot.show()
```
